```python
# Converting to GRAY
gr=cv2.cvtColor(pic,cv2.COLOR_BGR2GRAY)
```

```python
display(Image(data=cv2.imencode('.jpeg',gr)[1].tobytes()))
```

```python
# Invert
inv=cv2.bitwise_not(gr)
```

```python
display(Image(data=cv2.imencode('.jpeg',inv)[1].tobytes()))
```

```python
# Guassian Blur to Image
blur=cv2.GaussianBlur(inv,(17,15),0)
display(Image(data=cv2.imencode('.jpeg',blur)[1].tobytes()))
```

```python
#Blur to Inverted Blur
inbu=cv2.bitwise_not(blur)
display(Image(data=cv2.imencode('.jpeg',inbu)[1].tobytes()))
```

```python
#Creating Sketch
sketch=cv2.divide(gr,inbu,scale=250)
display(Image(data=cv2.imencode('.jpeg',sketch)[1].tobytes()))
```
